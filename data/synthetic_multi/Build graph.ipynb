```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
path = '/Users/francescoferrini/VScode/MultirelationalGNN/data/simple/tot_rel_5_metapaths_number_1'
```

```python
conv_node = {"100000":"blue",
 "010000":"red",
 "001000":"green",
 "000100":"purple",
 "000010":"orange",
 "000001":"yellow"}

conv_node = {"100":"blue",
 "010":"red",
 "001":"green",}
# "000100":"purple",
# "000010":"orange",
# "000001":"yellow"}

f = open(path + "/node.dat", "r")
nodes_emb = []
for x in f:
    a = x.split("\t")
    b = a[1][:-1].split(" ")
    b = "".join(b)
    
    tmp = [int(a[0]),conv_node[b]]
        
    nodes_emb.append(tmp)
```

```python
labels = pd.read_csv(path + "/label.dat",sep="\t",names=["n","l"]).to_numpy()
```

```python
pd.read_csv(path + "/metapath.dat",names=["a"]).a
```

```python
links = pd.read_csv(path + "/link.dat",names=["a","r","b"],sep="\t")
```

```python
edges = []
for a,b,c in links.to_numpy():
    edges.append((a,c,{"rel":b}))
    
g = nx.DiGraph()
g.add_edges_from(edges)

for n,e in nodes_emb:
    g.nodes()[n]["color"] = e
for n,l in labels:
    g.nodes()[n]["label"] = l
    
```

```python
n = 1
unique = dict()
unique2 = dict()

for n in g.nodes():
    node_col = g.nodes()[n]["color"]
    node_lab = g.nodes()[n]["label"]
    if node_lab == 1:
        for nei in list(nx.neighbors(g,n)):
            rel1 = g.edges()[(n,nei)]["rel"]
            nei_col = g.nodes()[nei]["color"]
            for nnei in list(nx.neighbors(g,nei)):
                rel2 = g.edges()[(nei,nnei)]["rel"]
                nnei_col = g.nodes()[nnei]["color"] 
                for nnnei in list(nx.neighbors(g,nnei)):
                    rel3 = g.edges()[(nnei,nnnei)]["rel"]
                    nnnei_col = g.nodes()[nnnei]["color"]
                    
                    for nnnnei in list(nx.neighbors(g,nnnei)):
                        rel4 = g.edges()[(nnnei,nnnnei)]["rel"]
                        nnnnei_col = g.nodes()[nnnnei]["color"]

                        #key = "LABEL="+str(node_lab)+" "+node_col+"-"+str(rel1)+"-"+nei_col+"-"+str(rel2)+"-"
                        key = "LABEL="+str(node_lab)+"       "+str(rel1)+"-"+nei_col+"-"+str(rel2)+"-"
                        kkey = key+nnei_col+"-"+str(rel3)+"-"+nnnei_col
                        key = kkey# + "-"+str(rel4)+"-"+nnnnei_col 
                        
                        
                        tmp = "LABEL="+str(node_lab)+"       "+str(rel1)+"-"+str(rel2)+"-"+str(rel3)#+"-"+str(rel4)
                        if key in unique:
                            unique[key]+= 1
                        else:
                            unique[key] = 1
                            
                        if tmp in unique2:
                            unique2[tmp]+= 1
                        else:
                            unique2[tmp] = 1
```
